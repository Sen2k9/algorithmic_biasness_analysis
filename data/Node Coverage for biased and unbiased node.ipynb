# Importing data for unbiased algorithm

```python
import pandas as pd
import seaborn as sns
tips2 = pd.read_csv("algo-bias.csv")
s = []
for each in tips2["Average node coverage(%)"]:
    a = str(each).split("%")
    s.append(int(a[0]))
tips2["Average node coverage(%)"]= s
op = []
for each in tips2["Optimal node coverage(%)"]:
    a = str(each).split("%")
    op.append(int(a[0]))
tips2["Optimal node coverage(%)"] = op
print(tips2.head())
```

# Ploting for Average node coverage for four metaheuristic algorithms (genetic algorithm, hill climbing with random walk, simulated annealing and tabu search)

# each having 200 iteration. The average node coverage = sum(optimal coverage for each generation)/200

# The figure shows that the average node coverage increases with the increase of node numbers

```python
#Average node
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
#sns.set_style("whitegrid")
plt.title('# of nodes vs Average node Coverage')
sns.scatterplot(x="# of nodes", y="Average node coverage(%)", data=tips2, marker='D',hue="Techniques",palette="deep")
plt.savefig('avg_coverage.jpg')
```

# The figure below is the boxplot representation of how four different algorithm varies for same node numbers and generation.

# For example, when number of nodes is 10 for 200 generation, the boxplot has lower bound at 40% and upper bound at 58% with 53% mean approximately.


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
#sns.set_style("whitegrid")
plt.title('# of nodes vs Average node Coverage')
sns.boxplot(
    data=tips2,
x="# of nodes", 
    y="Average node coverage(%)",
    color='red'
)
plt.savefig('avg_coverage_boxplot.jpg')
```

# The five figures below is the regression plot for average node coverage against number of nodes.

# The plot shows that when number of nodes get increased all the metaheuristic algorithms ( in our case four) tends to have positive linear relationship.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.title('# of nodes vs Optimal node Coverage')
sns.regplot(x="# of nodes", y="Average node coverage(%)", data=tips2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.title('# of nodes vs Average node Coverage')
sns.regplot(x="# of nodes", y="Average node coverage(%)", data=tips2, ci=90)
plt.savefig('avg_coverage_regplot.jpg')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.title('# of nodes vs Average node Coverage')
sns.regplot(x="# of nodes", y="Average node coverage(%)", data=tips2, x_bins=11)
plt.savefig('avg_coverage_regplot_x_bin.jpg')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
import numpy as np
sns.regplot(x="# of nodes", y="Average node coverage(%)", data=tips2, x_estimator=np.mean)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5))
import numpy as np
sns.lmplot(x="# of nodes", y="Average node coverage(%)", data=tips2, x_bins=11)
```
